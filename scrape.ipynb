```python
import selenium 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import joblib
import pandas as pd 
import numpy as np

```

```python
# initiallize the chrome driver
service=Service("D:/Desktop/imdb/imdb/chromedriver.exe")
```

```python

# Launch a new browser instance using Chrome
browser = webdriver.Chrome(service=service)

# Open IMDb's search page for feature films released in 2024
browser.get("https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31")
```

```python
while True:  # Infinite loop to keep clicking the "Load More" button until it no longer exists
    try:
        # Locate the "Load More" button using its class name
        load_element = browser.find_element(By.CLASS_NAME, "ipc-see-more__text")

        # Click the button to load more movies
        load_element.click()
    
    except:
        # If the button is not found, it means there are no more movies to load
        print("No more movies")
    
    break  # Exit the loop after the first attempt (remove this if you want continuous loading)
```

```python
# Find all movie titles (h3 tags) on the IMDb page
name = browser.find_elements(By.TAG_NAME, "h3")

# Find all movie story descriptions (using class name)
story = browser.find_elements(By.CLASS_NAME, "ipc-html-content-inner-div")
```

```python
fantansy = []  # Initialize an empty list to store dictionaries

for names, stories in zip(name, story):  # Iterate over 'name' and 'story' simultaneously
    fantansy.append({  # Append a dictionary to the list
        'name': names.text,  # Extract text from 'names' and store it under the key 'name'
        'story': stories.text  # Extract text from 'stories' and store it under the key 'story'
    })
```
